```julia
using BinBraKet
```

```julia
#Seven Orbits
Zero = zeros(8,1);
W=ket"001"+ket"010"+ket"100";
GHZ=ket"000"+ket"111";
b1=ket"0"⊗coev(2);
b2=(sigma(2,2)⊗id(2))∘(ket"0"⊗coev(2));
b3=coev(2)⊗ket"0";
pr=ket"0"⊗ket"0"⊗ket"0";

#epsilon tensor
ϵ= ket"0"∘bra"1" - ket"1"∘bra"0"
```

```julia
# Graphical Invariants
norm2(Ψ)= Ψ'∘Ψ; # ' is conjugate transpose
#bipartite invariants
bip1(Ψ)=Ψ'∘(id(2)⊗ϵ⊗ϵ)∘Ψ; 
bip2(Ψ)=Ψ'∘(ϵ⊗id(2)⊗ϵ)∘Ψ; 
bip3(Ψ)=Ψ'∘(ϵ⊗ϵ⊗id(2))∘Ψ; 
```

```julia
#Norm distinguishes the zero orbit
map(x->norm2(x)[1],(GHZ,W,b1,b2,b3,pr,Zero))
```

```julia
#bipartite invariants distinguish bipartite states from each other
[   bip1(b1) bip1(b2) bip1(b3);
    bip2(b1) bip2(b2) bip2(b3);
    bip3(b1) bip3(b2) bip3(b3)]
```

```julia
#Bipartite invariants nonzero on W and zero on GHZ
[bip1(W) bip2(W) bip3(W);
    bip1(GHZ) bip2(GHZ) bip3(GHZ)]
```

```julia
#for the hyperdeterminant we need swaps
σ=sigma(2,2); #use the default IntMat braiding
swap14 = (σ⊗σ)∘(id(2)⊗σ⊗id(2))∘(σ⊗σ);
hypdet(Ψ) = (Ψ'⊗Ψ')∘(id(2)⊗id(2)⊗swap14)∘(ϵ⊗ϵ⊗ϵ⊗ϵ⊗ϵ⊗ϵ)∘(Ψ⊗Ψ)
```

```julia
map(x->hypdet(x)[1],(GHZ,W,b1,b2,b3,pr,Zero))
```
